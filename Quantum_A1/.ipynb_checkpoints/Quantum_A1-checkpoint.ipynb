H gate applied to ket 1, making $\alpha$ = $\frac{1}{\sqrt{2}}$ and $\beta$ = $\frac{1}{\sqrt{2}}$. The probability of observing a 0 after measurement is 48.82813%, whereas observing is 1 is 51.17188%.

### P2


![Entanglement.png](Entanglement.png)  
Let the above circuit be called Circ:  
$Circ\ket{00} = \frac{1}{\sqrt{2}} (\, \ket{00} + \ket{11}\,)$, Upon measuring I get 48.82813% for $\ket{00}$ and 51.17188% for $\ket{11}$  
$Circ\ket{01} = \frac{1}{\sqrt{2}} (\, \ket{01} + \ket{10}\,)$, Upon measuring I get 48.82813% for $\ket{01}$ and 51.17188% for $\ket{10}$  
$Circ\ket{10} = \frac{1}{\sqrt{2}} (\, \ket{00} - \ket{11}\,)$, Upon measuring I get 48.82813% for $\ket{00}$ and 51.17188% for $\ket{11}$  
$Circ\ket{11} = \frac{1}{\sqrt{2}} (\, \ket{01} - \ket{10}\,)$, Upon measuring I get 48.82813% for $\ket{01}$ and 51.17188% for $\ket{10}$  
 

### P3

![Barriers.png](Barriers.png)  
At barrier 1 the state should be $\frac{1}{2}(\, \ket{00} + \ket{01} + \ket{10} + \ket{11}\,)$, which in words is just an equal superposition of all four computational basis states (ie, both qubits have an equal probability of being a $\ket{0}$ or a $\ket{1}$, and are not entangled with eachother).  
  
At barrier 2 the state should be $\frac{1}{2}(\, \ket{00} + \ket{01} + \ket{10} + \ket{11}\,)$, which is the same as above  
    
The final statevector after measurement reads as $\ket{00}$ with amplitude 1 and phase angle 0. I'm not entirely sure why that is the case, I will likely come to office hours to ask why (I think I understand that after measurement the statevector collapses to a single basis state, but I'm unsure as to why it collapsed to $\ket{00}$ specifically).

### P1

```qsharp
from qiskit import *
from qiskit.visualization import *

qc = QuantumCircuit()
qr = QuantumRegister(1,'q')
cr = ClassicalRegister(1,'creg')

qc.add_register(qr)
qc.add_register(cr)

qc.x(qr[0])
qc.z(qr[0])
qc.measure(qr[0], cr[0])
qc.draw()

```

```qsharp
qc = QuantumCircuit()
qr = QuantumRegister(1,'q')
cr = ClassicalRegister(1,'creg')

qc.add_register(qr)
qc.add_register(cr)

qc.x(qr[0])
qc.h(qr[0])
qc.measure(qr[0], cr[0])
qc.draw()
```

```qsharp
aer_sim = Aer.get_backend('aer_simulator')
job = aer_sim.run(qc, shots = 1024)
hist = job.result().get_counts()
plot_histogram(hist)
```

### P3

```qsharp
qc = QuantumCircuit()
qr = QuantumRegister(2,'q')
cr = ClassicalRegister(2,'creg')

qc.add_register(qr)
qc.add_register(cr)

qc.h(qr[1])
qc.h(qr[0])
qc.barrier(qr)
qc.cx(qr[0], qr[1])
qc.barrier(qr)
qc.measure(qr[0], cr[0])
qc.measure(qr[1], cr[1])
qc.draw()
```

```qsharp
operation P1Circuit() : Result {
    use q0 = Qubit();
    X(q0);
    Z(q0);
    let result = M(q0);
    Reset(q0);
    return result;
}
```

```qsharp
operation P1Circuit2() : Result {
    use q0 = Qubit();
    X(q0);
    H(q0);
    let result = M(q0);
    Reset(q0);
    return result;
}
```

```qsharp
open Microsoft.Quantum.Diagnostics;
operation P2Circuit1() : Result {
    use qs = Qubit[2];
    H(qs[0]);
    CNOT(qs[0], qs[1]);
    DumpMachine();
    let result = M(qs[0]);
    ResetAll(qs);
    return result;
}
```
